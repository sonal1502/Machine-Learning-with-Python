```python
#total bill vs tip datasets
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.count()
```

```python
sns.set_theme()
```

```python
#visualize dataset using plots

sns.relplot(data=tips,x='total_bill', y='tip', col='time',hue='smoker',style='smoker',size='size')
```
