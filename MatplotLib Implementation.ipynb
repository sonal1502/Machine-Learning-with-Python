```python
import matplotlib.pyplot as plt
#use numpy to take input of data
import numpy as np
x=np.linspace(0,10,100)  #10 equally spaced values betweej 0 and 10
y=np.sin(x) #linspce basically creates a line of values
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting data
plt.plot(x,y) #x - 100 values between 0 and 10, y - sin of all these values
plt.show()



```

```python
#to build cosine wave
plt.plot(x,z) # z - cosine value of x therefore we get a cosine wave
plt.show()
```

```python
#adding title to x and y axis of the figure
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine value")
plt.title("Sine wave")
plt.show()
```

```python
#to plot a parabola
x=np.linspace(-10,10,20) # gives us 20 equivalent values between -10 and +10
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x),'g-') #plots green continuous line
plt.plot(x, np.cos(x),'r--')#plots red dotted line
plt.show()
```

```python
#bar plot

fig=plt.figure()
ax= fig.add_axes([0,0,1,1]) #size of plot rectangle
languages= ["English","French","Spanish","Latin","German"]
people=[100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NO OF PEOPLE")
plt.show()
```

```python
fig2=plt.figure()
ax = fig2.add_axes([0,0,1,1]) #create plot space
languages= ["English","French","Spanish","Latin","German"]
people=[100,50,150,40,70]
ax.pie(people,labels=languages, autopct="%1.1f%%")
plt.show()

```

```python
x=np.linspace(0,10,20)
y=np.sin(x)
z=np.cos(x)

fig3= plt.figure()
ax =fig3.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig2=plt.figure()
ax = plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)

ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
